```python
import matplotlib.pyplot as plt
import seaborn as sns
for i in x.columns:
    sns.boxplot(x[i])
    plt.title(i)
    plt.show()
```

```python
import warnings 
warnings.filterwarnings("ignore")
for i in x.columns:
    print(i)
    Q1=x[[i]].quantile(0.25)
    Q2=x[[i]].quantile(0.5)
    Q3=x[[i]].quantile(0.75)
    print("Q1=",Q1[0])
    print("Q2=",Q2[0])
    print("Q3=",Q3[0])
    IQR=Q3[0]-Q1[0]
    print("IQR=",IQR)
    upp=Q3[0]+1.5*IQR
    low=Q1[0]-1.5*IQR
    print("LOW=",low)
    print("UPP=",upp)
    sns.boxplot(x[i])
    plt.title(i)
    plt.show()
    x1=x.loc[(x[i]>=low)&(x[i]<=upp)]
```

```python
for i in x1.columns:
    sns.histplot(x1[i],bins=20,kde=True)
    plt.title(i)
    plt.show()
```

```python
F=x.drop("Diagnosis",axis=1)
T=x["Diagnosis"]
```

```python
F.head()
```

```python
T
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(F,T,train_size=0.80,random_state=333)
```

```python
X_train.shape
```

```python
T.unique()
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
L=LogisticRegression()
from sklearn.model_selection import GridSearchCV
params={"C":[0.1,0.01,0.2,0.04,0.001],"penalty":["l1","l2"]}
log=GridSearchCV(L,param_grid=params,scoring="accuracy",cv=5)
```

```python
log.fit(X_train,y_train)
```

```python
log.best_params_
```

```python
model=log.best_estimator_
model
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
## SVC model is overfit finalize
```
