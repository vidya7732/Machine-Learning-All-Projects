```python
import warnings 
warnings.filterwarnings("ignore")
for i in x.columns:
    print(i)
    Q1=x[[i]].quantile(0.25)
    Q2=x[[i]].quantile(0.5)
    Q3=x[[i]].quantile(0.75)
    print("Q1=",Q1[0])
    print("Q2=",Q2[0])
    print("Q3=",Q3[0])
    IQR=Q3[0]-Q1[0]
    print("IQR=",IQR)
    upp=Q3[0]+1.5*IQR
    low=Q1[0]-1.5*IQR
    print("LOW=",low)
    print("UPP=",upp)
    sns.boxplot(x[i])
    plt.show()
    x1=x.loc[(x[i]>=low)&(x[i]<=upp)]
```

```python
F=x1.drop("quality",axis=1)
T=x1["quality"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(F,T,train_size=0.80,random_state=333)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
T.unique()
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,ComplementNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
models={"Gaussian":GaussianNB(),
        "Bernoulli":BernoulliNB(),
        "Complement":ComplementNB(),
        "SVC":SVC(),
        "KNN":KNeighborsClassifier(),
        "Logistic":LogisticRegression(),
       }
        
         
```

```python
models.keys()
```

```python
models.values()
```

```python
models.items()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
Data=[]
for i,j in models.items():
    j.fit(X_train,y_train)
    TR=j.score(X_train,y_train)
    TE=j.score(X_test,y_test)
    pred=j.predict(X_test)
    Data.append([i,TR,TE])
    print(i.upper())
    print(classification_report(y_test,pred))
    print(confusion_matrix(y_test,pred))
    print("_"*80)

```

```python
Data
```

```python
Result=pd.DataFrame(Data,columns=("models name","Training Accuracy","Testing Accuracy"))
Result
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(Result["models name"],Result["Training Accuracy"]*100,width=0.3,align="edge",label="TRAINING")
plt.bar(Result["models name"],Result["Testing Accuracy"]*100,width=0.3,align="center",label="TESTING")
plt.grid()
plt.legend()
plt.xlabel("models")
plt.ylabel("accuracy")
plt.title("model Training & Testing accuracy")
plt.show()
```
