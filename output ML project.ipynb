```python
x["Sex"]=x["Sex"].apply(lambda p:p.replace("M","1"))
x["Sex"]=x["Sex"].apply(lambda p:p.replace("F","0"))
x["Sex"]
```

```python
x["Anaemic"]=x["Anaemic"].map({"Yes":1,"No":0})
x["Anaemic"]
```

```python
F=x.drop("Anaemic",axis=1)
T=x["Anaemic"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(F,T,train_size=0.85,random_state=300)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
M=MinMaxScaler()
X_train[["%Red Pixel","%Green pixel","%Blue pixel","Hb"]]=M.fit_transform(X_train[["%Red Pixel","%Green pixel","%Blue pixel","Hb"]])
X_train
```

```python
X_test[["%Red Pixel","%Green pixel","%Blue pixel","Hb"]]=M.transform(X_test[["%Red Pixel","%Green pixel","%Blue pixel","Hb"]])
X_test
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
L=LogisticRegression()
from sklearn.model_selection import GridSearchCV
params={"C":[0.1,0.01,0.2,0.04,0.001],"penalty":["l1","l2"]}
log=GridSearchCV(L,param_grid=params,scoring="accuracy",cv=5)
```

```python
log.fit(X_train,y_train)
```

```python
log.best_params_
```

```python
log.best_estimator_
```

```python
log.score(X_train,y_train)
```

```python
log.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
s=SVC()
from sklearn.model_selection import GridSearchCV
params={"C":[.1,0.1,0.01,0.3],"gamma":[.1,.07,.01,1],"kernel":["rbf"]}
SVC=GridSearchCV(s,param_grid=params,scoring="accuracy",cv=7)
```

```python
SVC.fit(X_train,y_train)
```

```python
SVC.best_params_
```

```python
SVC.best_estimator_
```

```python
SVC.score(X_train,y_train)
```

```python
SVC.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
A=KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
knn=GridSearchCV(A,param_grid={"n_neighbors":[3,5,7,9]},scoring="accuracy",cv=5)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.best_params_
```

```python
knn.best_estimator_
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
A={"models":["log","SVC","knn"],"Train":[0.75,0.75,0.76],"Test":[0.75,0.75,0.68]}
```

```python
import matplotlib.pyplot as plt
plt.bar(A["models"],A["Train"],color="black",width=0.3,align="edge",label="Training Accuracy")
plt.bar(A["models"],A["Test"],color="red",width=0.3,align="center",label="Testing Accuracy")
plt.xlabel("models")
plt.ylabel("accuracy")
plt.title("Model Performance")
plt.legend()
plt.show()
```
